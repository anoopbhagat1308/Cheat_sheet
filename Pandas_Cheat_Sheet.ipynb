## Pandas Data Structures

Pandas has two primary data structures:
1. Series
2. DataFrame

### Series
A one-dimensional data structure.

```python
# creating series
import pandas as pd
import numpy as np

data = [10, 20, 30, 40]

s = pd.Series(data)

print(s)
```

### DataFrame

A DataFrame is a table consisting of rows and columns.

```python
data = {
    "Name": ["Alex", "Raj", "Priya", "Kim", "John", "Ron"],
    "Age": [35, 30, np.nan, 32, 29, 38],
    "Gender": ["M", "M", "F", "F", "M", "M"],
    "Region": [" ", "South", "North", " ", "East", "East"],
    "Sales": [1000, 5000, 2500, 3000, 2000, 5500]
}

df = pd.DataFrame(data)

display(df)
```
